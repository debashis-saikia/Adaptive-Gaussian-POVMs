```python
from GaussianKraus_Kmeans import QuantumSegmentation as QS

image = "/content/The-256-256-Lena-image.jpeg" # Changed image path to an available image in Colab
QS = QS(image)
QS.run_segmentation(k_clusters=3)
metrics = QS.calculate_metrics()
print(f"PSNR: {metrics['PSNR']:.2f} dB")
print(f"SSIM: {metrics['SSIM']:.4f}")
print(f"Entropy Change: {metrics['Entropy Change']:.4f}")
print(f"Percent Change in Entropy: {metrics['Percent Change']:.2f}%")
QS.plot_images()
QS.plot_results()
```

```python
from GaussianKraus_Kmeans import QuantumSegmentation as QS

# List of the specific WhatsApp images provided
image_paths = [
    "/content/WhatsApp Image 2026-03-14 at 9.38.16 PM.jpeg",
    "/content/WhatsApp Image 2026-03-14 at 9.38.17 PM (1).jpeg"
]

for path in image_paths:
    print(f"--- Processing Image: {path} ---")
    # Initialize and run segmentation
    qs_instance = QS(path)
    qs_instance.run_segmentation(k_clusters=3)

    # Calculate and display metrics
    results = qs_instance.calculate_metrics()
    print(f"PSNR: {results['PSNR']:.2f} dB")
    print(f"SSIM: {results['SSIM']:.4f}")
    print(f"Entropy Change: {results['Entropy Change']:.4f}")
    print(f"Percent Change: {results['Percent Change']:.2f}%")

    # Display plots
    qs_instance.plot_images()
    qs_instance.plot_results()
    print("\n" + "="*30 + "\n")
```

```python
import matplotlib.pyplot as plt
from google.colab import files
import os
from GaussianKraus_Kmeans import QuantumSegmentation as QS

# List of all images processed so far
all_images = [
    ("/content/The-256-256-Lena-image.jpeg", "lena"),
    ("/content/WhatsApp Image 2026-03-14 at 9.38.16 PM.jpeg", "whatsapp_1"),
    ("/content/WhatsApp Image 2026-03-14 at 9.38.17 PM (1).jpeg", "whatsapp_2")
]

saved_files = []

for path, label in all_images:
    if os.path.exists(path):
        print(f"Processing and saving results for: {label}")
        qs_instance = QS(path)
        qs_instance.run_segmentation(k_clusters=3)

        # Save Segmentation Comparison Plot
        qs_instance.plot_images()
        segment_filename = f"{label}_segmentation.png"
        plt.savefig(segment_filename)
        plt.close()
        saved_files.append(segment_filename)

        # Save Results/Metrics Plot
        qs_instance.plot_results()
        metrics_filename = f"{label}_metrics.png"
        plt.savefig(metrics_filename)
        plt.close()
        saved_files.append(metrics_filename)

# Download all saved files
print(f"Downloading {len(saved_files)} files...")
for f in saved_files:
    files.download(f)
```
